```python
# Нужно промоделировать какое-то теоретическое распределение и сравнить его с 
# эмпирическим при разном числе элементов выборки. 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def norm_th(a, sigma):
    x = np.linspace(a-3*sigma, a+3*sigma, 1000)  # массив точек по оси x
    y = norm.cdf(x, a, sigma)  # массив точек вдоль оси y
    plt.plot(x, y) #отрисовка
    plt.xlim([a-5*sigma, a+5*sigma])  # границы оси x
    plt.ylim([0,1]) # границы оси y  
    
def emp(n, a, sigma):  # n - количество элементов в выборке
    samples = np.random.normal(a, sigma, n)  # выборка из нормального распределения из n элементов
    x = np.sort(samples)  # вариационный ряд
    y = np.arange(0, len(samples)) / len(samples)  # точки по оси y
    for i in range(len(y)):  
        plt.plot(x[i], y[i],  marker='o')  # отрисовка точек
    plt.xlim([a-5*sigma, a+5*sigma])  # границы оси x
    plt.ylim([0,1]) # границы оси y 

    
norm_th(1,2) # возьмем нормальное рапспределение с параметрами 1,2
emp(10, 1, 2) # выборка из 10 элементов

plt.show()
```

```python
emp(100,1,2) # выборка из 100 элементов
norm_th(1,2)
plt.show()
```

```python
emp(1000,1,2) # выборка из тысячи элементов
norm_th(1,2)
plt.show()
```
